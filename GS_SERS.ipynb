```python
import random
import pandas as pd
import matplotlib.pyplot as plt

#simulação dos dados da missão
temperatura = random.randint(0, 100)
energia_solar = random.randint(0, 100)
energia_bateria = random.randint(0,100)
comunicacao = random.choice(["online", "offline"])
modulo_operacao = random.choice(["online", "offline"])

#exibição de dados
print("MONITORAMENTO DA MISSÃO ESPACIAL")
print(f"Temperatura: {temperatura} °C")
print(f"Energia da bateria: {energia_bateria}%")
print(f"Geração de energia solar: {energia_solar}%")
print(f"Comunicação: {comunicacao}")
print(f"Módulo de operação: {modulo_operacao}")

#alertas automáticos
print("\nALERTAS")

if temperatura > 80:
    print("Alerta: Temperatura crítica!")

if energia_bateria < 30:
    print("Alerta: Nível de bateria baixo!")

if comunicacao == "offline":
    print("Alerta: Falha na comunicação!")

if energia_solar < 20:
    print("Alerta: Baixa geração de energia solar!")

if modulo_operacao== "offline":
    print("Alerta: Módulo de operação desligado!")

#tomada de decisão automática
print("\nDECISÕES AUTOMÁTICAS")

if energia_bateria < 30:
    print("Modo economia ativado.")
    print("Desligando operações não essenciais.")

if energia_solar > 60:
    print("Energia solar suficiente.")
    print("Prioridade para uso de energia renovável.")

if temperatura > 80:
    print("Sistema de resfriamento ativado.")

if comunicacao == "offline":
    print("Tentando restabelecer comunicação.")

#organização de dados
dados = {
    "Temperatura": [temperatura],
    "Bateria": [energia_bateria],
    "Energia Solar": [energia_solar],
    "Comunicação": [comunicacao],
    "Módulo de operação": [modulo_operacao]
}

tabela = pd.DataFrame(dados)

print("\n DADOS ORGANIZADOS")
print(tabela.to_string(index=False))

#visualização de dados
plt.figure(figsize=(6,4))

valores = [temperatura, energia_bateria, energia_solar]
nomes = ["Temperatura", "Bateria", "Energia Solar"]

plt.bar(nomes, valores)

plt.title("Monitoramento Energético da Missão")
plt.ylabel("Valores")

plt.show()

#sustentabilidade
print("\nSUSTENTABILIDADE")
print("O sistema busca reduzir desperdícios energéticos e priorizar fontes sustentáveis de energia.")
if energia_solar > energia_bateria:
    print("Perfeito! A missão está utilizando principalmente energia renovável!")
else:
    print("Infelizmente, a missão está consumindo mais energia armazenada da bateria.")
```
